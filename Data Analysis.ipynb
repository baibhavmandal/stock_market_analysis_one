## Closing Price
The closing price is the last price at which the stock is traded during the regular trading day. A stock’s closing price is the standard benchmark used by investors to track its performance over time.

1. subplots_adjust() - This function allows you to fine-tune the layout of subplots within a figure. It provides control over the spacing between subplots, including the height and width of the space between them.

2. enumerate() - The enumerate() method is a useful tool in Python for iterating through elements in an iterable (like a list or tuple) while keeping track of the index of the current element. It returns an enumerate object that contains pairs of index and value for each element in the iterable.

3. tight_layout() - The tight_layout() function is a convenient way to automatically adjust the subplot parameters to ensure that all subplots fit neatly into the figure area. It eliminates overlapping and avoids clipping of axis labels or titles.

Also, you can refer to the provided link for understanding how to plot a pandas DataFrame: https://pandas.pydata.org/docs/getting_started/intro_tutorials/04_plotting.html

```python
# Let's see a historical view of the closing price
plt.figure(figsize=(15, 10))
plt.subplots_adjust(top=1.25, bottom=1.2)

for j, company in enumerate(company_list, 1):
    plt.subplot(3, 2, j)
    company['Adj Close'].plot()
    plt.ylabel('Adj Close')
    plt.xlabel(None)
    plt.title(f"Closing Price of {tech_list[j - 1]}")
    
plt.tight_layout()
```

## What was the moving average of the various stocks?
The moving average (MA) is a simple technical analysis tool that smooths out price data by creating a constantly updated average price. The average is taken over a specific period of time, like 10 days, 20 minutes, 30 weeks, or any time period the trader chooses.

Here is a link to a study about the rolling() and mean() methods in Python pandas DataFrame - https://www.geeksforgeeks.org/python-pandas-dataframe-rolling/. This resource explains how to use the rolling() method, which enables you to perform rolling window calculations, and the mean() method, which calculates the mean of the values within the window.

Here are some useful links to study more about subplots:
1. The following link provides a demonstration and examples of working with subplots in Matplotlib - https://matplotlib.org/stable/gallery/subplots_axes_and_figures/subplots_demo.html. This resource showcases how to create and customize subplots in a clear and visual manner.
2. Additionally, this Stack Overflow link sheds light on the "ax" parameter in the pandas Series plot - https://stackoverflow.com/questions/64545066/make-sense-of-parameter-ax-in-pandas-series-plot#:~:text=By%20providing%20a%20plotting%20function,assumes%22%20you%20want%20it%20to. The explanation clarifies how passing the "ax" parameter allows you to have more control over the plotting process by specifying the axes on which the plot will be drawn.


```python
ma_day = [10, 20, 50]

for ma in ma_day:
    for company in company_list:
        column_name = f"MA for {ma} days"
        company[column_name] = company['Adj Close'].rolling(ma).mean()
        

fig, axes = plt.subplots(nrows=3, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(15)

TCS[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[0,0])
axes[0,0].set_title('Tata')

HCLTECH[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[0,1])
axes[0,1].set_title('HCL')

INFY[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[1,0])
axes[1,0].set_title('Infosys')

TECHM[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[1,1])
axes[1,1].set_title('Mahindra')

WIT[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[2,0])
axes[2,0].set_title('Wipro')

AAPL[['Adj Close', 'MA for 10 days', 'MA for 20 days', 'MA for 50 days']].plot(ax=axes[2,1])
axes[2,1].set_title('Apple')

fig.tight_layout()
```

We see in the graph that the best values to measure the moving average are 10 days because we still capture trends in the data without noise.

##  What was the daily return of the stock on average?
Now that we've done some baseline analysis, let's go ahead and dive a little deeper. We're now going to analyze the risk of the stock. In order to do so we'll need to take a closer look at the daily changes of the stock, and not just its absolute value. Let's go ahead and use pandas to retrieve teh daily returns for the Apple stock.

pct_change() - This method calculates the percentage change in the values through a series. For example, if the values are [2,3,6] , the returned values would be [NaN, 0.5, 1.0] . This is because there is a 50% increase from the first element to the second and a 100% increase from the second element to the third.

```python
# We'll use pct_change to find the percent change for each day
for company in company_list:
    company['Daily Return'] = company['Adj Close'].pct_change()
    
fig, axes = plt.subplots(nrows=3, ncols=2)
fig.set_figheight(10)
fig.set_figwidth(15)

TCS['Daily Return'].plot(ax=axes[0,0])
axes[0,0].set_title('Tata')

HCLTECH['Daily Return'].plot(ax=axes[0,1])
axes[0,1].set_title('HCL')

INFY['Daily Return'].plot(ax=axes[1,0])
axes[1,0].set_title('Infosys')

TECHM['Daily Return'].plot(ax=axes[1,1])
axes[1,1].set_title('Mahindra')

WIT['Daily Return'].plot(ax=axes[2,0])
axes[2,0].set_title('Wipro')

AAPL['Daily Return'].plot(ax=axes[2,1])
axes[2,1].set_title('Apple')

fig.tight_layout()
```

Great, now let's get an overall look at the average daily return using a histogram. We'll use seaborn to create both a histogram and kde plot on the same figure.

```python
plt.figure(figsize=(12, 9))

for j, company in enumerate(company_list, 1):
    plt.subplot(3, 2, j)
    company['Daily Return'].hist(bins=50)
    plt.xlabel('Daily Return')
    plt.ylabel('Counts')
    plt.title(f'{company_name[j - 1]}')
    
plt.tight_layout()
```

While the simplicity of just calling sns.pairplot() is fantastic we can also use sns.PairGrid() for full control of the figure, including what kind of plots go in the diagonal, the upper triangle, and the lower triangle. Below is an example of utilizing the full power of seaborn to achieve this result.

```python
# Set up our figure by naming it returns_fig, call PairPLot on the DataFrame
return_fig = sns.PairGrid(tech_rets.dropna())

# Using map_upper we can specify what the upper triangle will look like.
return_fig.map_upper(plt.scatter, color='purple')

# We can also define the lower triangle in the figure, inclufing the plot type (kde) 
# or the color map (BluePurple)
return_fig.map_lower(sns.kdeplot, cmap='cool_d')

# Finally we'll define the diagonal as a series of histogram plots of the daily return
return_fig.map_diag(plt.hist, bins=30)
```

```python
# Set up our figure by naming it returns_fig, call PairPLot on the DataFrame
returns_fig = sns.PairGrid(closing_df)

# Using map_upper we can specify what the upper triangle will look like.
returns_fig.map_upper(plt.scatter,color='purple')

# We can also define the lower triangle in the figure, inclufing the plot type (kde) or the color map (BluePurple)
returns_fig.map_lower(sns.kdeplot,cmap='cool_d')

# Finally we'll define the diagonal as a series of histogram plots of the daily return
returns_fig.map_diag(plt.hist,bins=30)
```

## How much value do we put at risk by investing in a particular stock?
There are many ways we can quantify risk, one of the most basic ways using the information we've gathered on daily percentage returns is by comparing the expected return with the standard deviation of the daily returns.

```python
rets = tech_rets.dropna()

rets.head()
```

```python
area = np.pi * 20

plt.figure(figsize=(10, 8))
plt.scatter(rets.mean(), rets.std(), s=area)
plt.xlabel('Expected return')
plt.ylabel('Risk')

for label, x, y in zip(rets.columns, rets.mean(), rets.std()):
    plt.annotate(label, xy=(x, y), xytext=(50, 50), textcoords='offset points', ha='right', va='bottom', 
                 arrowprops=dict(arrowstyle='-', color='blue', connectionstyle='arc3,rad=-0.3'))
```
